```python
#K-Fold 교차검증
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

from sklearn.model_selection import KFold

ㅁㅁ_data = ㅁㅁ.data
ㅁㅁ_label = ㅁㅁ.target
fold_df_clf = DecisionTreeClassifier()

#폴드 세트를 5개로 분리하고 세트별 정확도를 담을 리스트 생성
kfold = KFold(n_splits=5)
cv_accuracy = []

#시작

n_iter = 0
for train_idx, test_idx in kfold.split(ㅁㅁ_data) :
    X_train, X_test = ㅁㅁ_data[train_idx], ㅁㅁ_data[test_idx]
    
    y_train, y_test = ㅁㅁ_label[train_idx], ㅁㅁ_label[test_idx]
    
    #학습 진행
    fold_df_clf.fit(X_train, y_train)
    
    #예측
    fold_pred = fold_df_clf.predict(X_test)
    
    #정확도 측정
    n_iter += 1
    accuracy = np.round(accuracy_score(y_test, fold_pred), 4)
    print('\n{} 교차검증 정확도 : {}, 학습 데이터 크기 : {}, 검증 데이터 크기 : {}'.format(n_iter, accuracy, X_train.shape[0], X_test.shape[0]))
    
    cv_accuracy.append(accuracy)
print('\n')

print('\n 평균검증 정확도 : ', np.mean(cv_accuracy))
```

```python
#손실 함수(MSE), 회귀지표(MAE)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.model_selection import train_test_split

x_data=scaled_df.loc[:, ['쓸 칼럼1', '쓸 칼럼2']]
y_data=scaled_df.loc[:, '타깃']
X_train, X_test, y_train, y_test=train_test_split(x_data,
                                                  y_data,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)


lr = LinearRegression()

lr.fit(X_train, y_train)


print("회귀계수(기울기):", np.round(lr.coef_, 1)) #np: numpy, coef_: 피처에 대한 회귀 계수 값
print("상수항(절편):", np.round(lr.intercept_, 1))  #intercept_: 상수항(절편) 값


y_preds = lr.predict(X_test)
print(X_train.shape, X_test.shape)


mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)
print(f'MSE : {mse:.3f}, RMSE: {rmse:.3f}')
print(f'Variance score : {r2_score(y_test, y_preds):.3f}')


print("절편 값:", lr.intercept_) # y축 절편값

# 회귀 계수(coefficient) : 독립변수의 변화에 따라 종속변수에 미치는 영향력이 크기
print("회귀계수:", np.round(lr.coef_,1))


plt.figure(figsize=(10,5))
plt.scatter(X_test['쓸 칼럼1'], y_test, label='y_test')
plt.scatter(X_test['쓸 칼럼1'], y_preds, c='r', label='y_pred')
#plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show()           

```
